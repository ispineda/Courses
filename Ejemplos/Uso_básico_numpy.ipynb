```python
import matplotlib as plt
import matplotlib.pyplot as plt
import numpy as np
from scipy import misc
import urllib.request
from PIL import Image
from matplotlib.pyplot import figure

plt.style.use("dark_background")
```

```python
url = "https://www.muycomputer.com/wp-content/uploads/2021/08/Spider-Man.jpg"
internet_image = Image.open(urllib.request.urlopen(url))
```

```python
image1 = np.array(internet_image)
image2 = np.array(internet_image)
image3 = np.array(internet_image)
```

```python
plt.figure(figsize=(10,20))
plt.imshow(image1)
```

```python
print(image1.shape)
```

```python
# Valor m√≠nimo y valor m√°ximo

print(np.amin(image1), np.amax(image1))
```

```python
#Impresi√≥n de filas

print(image1[150])
print(image1[150].shape)
```

```python
image2[100:500, :, 0] = 255
plt.figure(figsize=(10,20))
plt.imshow(image2)
```

```python
image2[:100, : , 1] = 255
plt.figure(figsize=(10,20))
plt.imshow(image2)
```

```python
image2[500:, : , 2] = 255
plt.figure(figsize=(10,20))
plt.imshow(image2)
```

```python
# Filtros
filtro = image3 > 150
print(filtro.shape)
print(image3.shape)
```

```python
image3[filtro] = 0
plt.figure(figsize=(10,20))
plt.imshow(image3)
```

```python
print(image1.shape)
rows, cols, layers = image1.shape
center_y = int(rows/2)
center_x = int(cols/2)
print(center_y, center_x)

# Filtro primer imagen
Y, X = np.ogrid[:rows, :cols]
```

```python
dist_from_center = (X-center_x)**2 + (Y-center_y)**2
print(dist_from_center)
print(np.amax(dist_from_center))
```

```python
radius = center_y**2
print(radius)
```

```python
circular_mask = (dist_from_center > radius)
image1[circular_mask] = 0
plt.figure(figsize=(10,20))
plt.imshow(image1)
```
